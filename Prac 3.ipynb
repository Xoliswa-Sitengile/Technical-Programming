```python
import pandas as pd 
import numpy as np 
import seaborn as sns
import matplotlib.pyplot as plt
```

Task 1

```python
exams = pd.read_csv('exams (2).csv')
exams.head(2)
```

Task 2

```python
#Create a box plot for math, reading & writing scores
exams[['math score', 'reading score', 'writing score']].plot(kind='box')
plt.title("Box plot of math, reading & writing scores")
plt.ylabel("Scores")
plt.show()
```

Task 3

```python
#Create a scatter plot reading & math score
exams.plot(kind='scatter', x='math score', y='reading score', color='blue')
plt.title('Scatter plot for Math and Reading Scores')
plt.xlabel("Math score")
plt.ylabel("Reading score")
plt.show() 
```

Question 1

```python
#Calculate correlation
correlation = exams['reading score'].corr(exams['writing score'])
correlation
```

Question 2

```python
#Calcualte the mean score for each lunch type
lunch_mean = exams.groupby('lunch')[['math score','reading score','writing score']].mean()
print(lunch_mean)

#Plot the average scores by lunch type
lunch_mean.plot(kind='bar')
plt.title("Average tests scores by lunch type")
plt.xlabel("Lunch Type")
plt.ylabel("Average Score")
plt.show()
```
