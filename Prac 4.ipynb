```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Task 1

```python
df_exams = pd.read_csv("exams (2).csv")
df_exams.head()
```

Task 2

```python
sns.catplot(x='parental level of education', kind='count',data=df_exams)
plt.show()
```

Task 3

```python
#Rotate the x labels
sns.catplot(x='parental level of education', kind='count',data=df_exams)
plt.xticks(rotation=45)
plt.show()
```

Task 4

```python
sns.scatterplot(x='writing score', y='reading score', hue ='test preparation course', data=df_exams)
plt.show()
```

Task 5

```python
sns.relplot(x='writing score', y='reading score', hue ='test preparation course', data=df_exams)
plt.show()
```

Task 6

```python
sns.relplot(x='writing score', y='reading score', hue ='test preparation course', data=df_exams, height=6, aspect=1.5)
plt.show()
```
